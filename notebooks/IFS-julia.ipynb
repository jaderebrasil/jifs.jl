### Exemplo 3 (testing QuadPlot)

```julia
include("../src/ifs.jl")
num_points = 1000
iters = 10000
halfmap = Jifs.IFSMap(x -> x ./ 2, color=[1., 0., 0.])
img = Jifs.ImageData(10, 10)

ifs = Jifs.IFS(img, [halfmap], [1])
Jifs.painting!(ifs, iters, num_points, quad=Jifs.Quad0)
```

```julia
Jifs.showifs(ifs, iters)
```
